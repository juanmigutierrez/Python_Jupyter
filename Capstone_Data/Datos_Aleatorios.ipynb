Es necesario cuando generamos las notas con aleatoriedad, que es posible que nos de valores mayores a 5 y menores a 0, por la distribución normal que utilizamos. Si se pasa de 5, se pone 5 la nota y si pasa por debajo de 0, la nota es 0.

```python
def corregir(nota):
    if nota >5:
        nota = 5
    elif nota <0:
        nota = 0
    nota = round(nota,1) #Redondeo un decimal de la nota
    return nota
```

```python
def ano_academ(cohorte,semestre_entro,semestre):
    # cohorte: año en el que entra el estudiante al pregrado
    # semestre_entro: semestre 1 o 2 del año en el que entra el estudiante al pregrado
    # semestre: semestre que está viendo el estudiante
    # retorna variables de ano_academico y semestre academico

    if semestre_entro == 1:
        ano_academ = cohorte
    else:
        ano_academ = cohorte + 0.5

    ano_academ = ano_academ+(semestre-1)*0.5

    if ano_academ % 1 == 0:
        cohorte_academ = 1
        ano_academ = math.floor(ano_academ)
    else:
        cohorte_academ = 2
        ano_academ = math.floor(ano_academ)

    return ano_academ, cohorte_academ
```

# Generar estudiante antiguo
Para generar un estudiante antiguo se llama a la función generar_estudiante2 de tal forma que para un semestre determinado genere todas sus notas en los semestres pasados

```python
def generar_estudiante_antiguo(df, accesos, subaccesos,sexo, cohorte, semestre_entro,semestre,media,varianza):
    semestre_t = semestre
    df_t = df
    registro = generar_estudiante1(accesos, subaccesos, sexo)
    while semestre_t > 0 :
        df_t = generar_estudiante2(registro,df_t, accesos, subaccesos, sexo, cohorte, semestre_entro,semestre_t,media,varianza)
        semestre_t = semestre_t -1
    return df_t
```

```python
cortes = [corte1,corte2,corte3,corte4,corte5]
porcentajes =[porcentaje1,porcentaje2,porcentaje3,porcentaje4,porcentaje5]
columnas = ['Documento Ingreso','Documento Actual','Codigo','Apellido 1 Estudiante','Apellido 2 Estudiante','Nombres Estudiante','Acceso','Subacceso','Correo Institucional','Sexo','Facultad o Escuela','Programa','Semestre','Creditos inscritos','Codigo Asignatura','Nombre Asignatura','Creditos Asignatura','Grupo Asignatura','Tipologia Asignatura','Nota 1er Corte','Corte_1p','Nota 2do Corte','Corte_2p','Nota 3er Corte','Corte_3p','Nota 4to Corte','Corte_4p','Nota 5to Corte','Corte_5p','Alerta']
df_alerta =pd.DataFrame(columns = columnas)
def cortes_acumulados(inicial,final,df):
    # inicial: corte inicial para el promedio acumulado (numero)
    # final: corte final para el promedio acumulado (numero)
    acumulado =[]
    for j in asignaturas:
        # primero se recorre todas las asignaturas
            materia = df[df["Nombre_Asignatura"]== j]
            for index,row in materia.iterrows():
                num = 0
                den = 0
                for i in range(inicial-1,final):
                    # Luego recorremos los cortes especificados
                    num = num + row[cortes[i]]*row[porcentajes[i]]
                    den = den + row[porcentajes[i]]
                    prom_acum = num/den
                if prom_acum >= 2 and prom_acum < 3:
                        print("El Estudiante "+ row["Nombres_Estudiante"]+
                          " se encuentra en Alerta nota baja, en el corte acumulado ("+str(inicial)+","+str(final)+") en la materia " + j )
                elif prom_acum < 2:
                        print("El Estudiante "+ row["Nombres_Estudiante"]+
                              " se encuentra en Alerta nota MUY baja, en el corte acumulado ("+str(inicial)+","+str(final) +") en la materia " + j )

```

```python
# Funcion que modifica las notas para una materia
def modificar_mate(materia,corte,media,varianza,data=None):
    # nombre_materia: nombre de la asignatura
    # corte : string de corte respectivo
    # media : media
    # varianza : varianza
    if data is None: # Se utiliza df completo si no se l e pasan datos
        notas=[]
        asignatura = df_completo[df_completo["Nombre_Asignatura"]== materia]
        largo = asignatura.shape[0] # se obtiene cuantos estudiantes la estan viendo, la materia
        for i in range(largo):
            nota = random.normalvariate(media,varianza) # le modificados a un estudiante la nota con la media y varianza respectiva que recibe la funcion
            nota = corregir(nota)
            notas.append(nota)
        asignatura[corte] = notas
    else:
        notas=[]
        asignatura = data
        largo = asignatura.shape[0]
        for i in range(largo):
            nota = random.normalvariate(media,varianza)
            nota = corregir(nota)
            notas.append(nota)
        asignatura[corte] = notas
    return asignatura
```

```python
# Funcion que modifica las notas por corte
def modificar_cortes_mate(materia,corte_i,corte_f,media,varianza):
    asignatura = modificar_mate(materia,corte_i,media,varianza) #Inicializar asignatura
    for i in range(corte_i -1,corte_f): # modifica cada corte de la amteria
        corte = cortes[i]
        asignatura = modificar_mate(materia,corte,media,varianza,asignatura)
    return asignatura
```

```python
# Funcion  que levanta alerta para materias por corte
def alerta_materia_corte(nombre_materia,num_corte,df):
    # nombre_materia: nombre de la asigntura
    # num_corte: numero del corte del semestre (de 1 a 5)
    # df: dataframe sobre el cual se hace el análisis
    asignatura = df[df["Nombre_Asignatura"]==nombre_materia]
    prom_corte = asignatura[cortes[num_corte-1]].mean()
    #print("Promedio en el corte " + str(num_corte))
    #print(prom_corte)
    if prom_corte >= 2 and prom_corte < 3:
        print("El promedio de "+nombre_materia+" en el corte "+str(num_corte)+" es BAJO")
    elif prom_corte < 2:
        print("El promedio de "+nombre_materia+" en el corte"+str(num_corte)+" es MUY BAJO")
```

```python
# Funcion que levanta alerta para materias por promedio acumulado
def alerta_materias_acumulado(nombre_materia,inicial,final,df):
    # nombre_materia: nombre de la asignatura
    # inicial: numero del corte inicial sobre el cual se calcula el promedio acumulado
    # final: numero del corte final sobre el cual se calcula el promedio acumulado
    # df: dataframe dobre el cual se hace el analisis
    num = 0
    den = 0
    asignatura = df[df["Nombre_Asignatura"]==nombre_materia]
    #print(asignatura[porcentajes[0]].mean())
    for i in range(inicial-1,final):
        peso = asignatura[porcentajes[i]].mean()
        num = num + asignatura[cortes[i]].mean()*peso
        den = den + peso
    prom_acum_mat = num/den
    #print("promedio acumulado: ",prom_acum_mat)
    if prom_acum_mat >= 2 and prom_acum_mat < 3:
        print("El promedio acumulado de "+str(inicial)+","+str(final)+" "+nombre_materia+" es BAJO")
    elif prom_acum_mat < 2:
        print("El promedio acumulado de "+str(inicial)+","+str(final)+" "+nombre_materia+" es MUY BAJO")
```
