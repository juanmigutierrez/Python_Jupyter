## Problema de Buscaminas

### Objetivo

Continuamos con el enfoque basado en agentes de la Inteligencia Artificial. Ahora usaremos la representación de conocimiento para construir programas de agentes más sofisticados. En este notebook y el siguiente usaremos la lógica proposicional para mantener una base de conocimiento y un estado de creencias que permita a un agente superar el reto de despejar todo el tablero (sin morir en el intento).

La **definición formal del entorno** se hace con base en la definición de las siguientes características:

* **Entorno**: Un mapa representado por una rejilla $n\times n$ bordeada por muros. El agente siempre comienza con todas las casillas tapadas mirando a la derecha. Comienza abriendo una casilla al azar. El numero de bombas en las diferentes casillas equivale al techo del 20 % de las casillas totales.

* **Actuadores**: El jugador puede elegir destapar una casilla o marcar bomba (Poner bandera).

* **Sensores**: El jugador puede percibir todas las casillas que ha decidido destapar y el numero de bombas de banderas que le quedan y cuantas bombas existen en el mapa.

* **Medida de desempeño**: No existe medida desempeño para el juego.

**Implementación del entorno**

Implementaremos el mundo del Buscaminas mediante una clase en Python.

Primero cargar librerías y definir unas funciones auxiliares útiles:

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import random as rd
import math as mt
import numpy as np
from PIL import Image
from itertools import combinations
```

```python
def truncar(x,p):
    if x < 0:
        return 0
    elif x > p-1:
        return p-1
    else:
        return x

def adyacentes(casilla,n,m):
    x, y = casilla
    adyacentes = [
        (truncar(x - 1,n), y), (truncar(x + 1,n), y),
        (x, truncar(y - 1,m)), (x, truncar(y + 1,m)),
        #
        (truncar(x - 1,n), truncar(y + 1,m)), (truncar(x + 1,n), truncar(y - 1,m)),
        (truncar(x - 1,n), truncar(y - 1,m)), (truncar(x + 1,n), truncar(y + 1,m)),
        #
    ]
    adyacentes = [c for c in adyacentes if c != casilla]
    adyacentes = list(dict.fromkeys(adyacentes))
    return adyacentes

print(adyacentes((3,3),5,5))
```

Las casillas se encuentran codificadas de la siguiente manera:

* si $0 \leq n \leq 8$ representa el numero de bombas $n$ adyacentes a esa casilla.

* **9**: si hay una bomba

* **10** Casilla Tapada

* **11** : Bandera sobre la casilla

```python

class buscaminas:
    def __init__(self,n,m):
        self.estado=[] # Lo que el jugador ve
        self.mapa=[] #El mapa completo
        self.n=n ##numero de filas
        self.m=m #numero de columnas
        self.bombas = [] #indices de las bombas
        self.juego_activo = False
        self.num_banderas = 0;

    def estado_inicial(self):
        self.juego_activo = True
        self.estado = np.full((self.n,self.m),10) #Crea matriz nxm llena de 10s
        self.mapa = np.zeros((self.n,self.m)) #Crea matriz nxm llena de 0s inicialmente, despues se va llenando
        num_bombas=mt.ceil(self.n*self.m*0.2)
        self.num_banderas = num_bombas
        # Poner Bombas
        for i in range(num_bombas):
            aleatorio1 = rd.randint(0,self.n-1) #Genera aleatorio
            aleatorio2 = rd.randint(0,self.m-1) #Genera aleatorio
            while ((aleatorio1,aleatorio2) in self.bombas): # Si no hay bombas genera bomba
                aleatorio1 = rd.randint(0,self.n-1)
                aleatorio2 = rd.randint(0,self.m-1)
            self.bombas.append((aleatorio1,aleatorio2))
            self.mapa[aleatorio1,aleatorio2] = 9
        #Poner numeros en las casillas faltantes
        for i in range(self.n):
            for j in range(self.m):
                if not self.mapa[i,j] == 9:
                    bombas = 0
                    ady = adyacentes((i,j),self.n,self.m) #obtiene casilals adyacentes
                    for adys in ady:
                        if self.mapa[adys] == 9: # reconocio bomba en una casilla adyacente
                            bombas = bombas +1
                    self.mapa[i,j] = int(bombas) #llena la casilla con el numero de bombas adyacentes
        
        
    def pintar_tablero(self):
        # Presenta gráficamente el tablero que el jugador puede ver
        fig,ax = plt.subplots(self.n,self.m,figsize=(2,2))
        plt.subplots_adjust(wspace=0, hspace=0) # unes las figuras
        for i in range(self.n):
            for j in range(self.m):
                with open("./Images/{}.JPG".format(int(self.mapa[i,j])), 'rb') as f:
                    img = Image.open(f)
                    plt.axis('on')
                    ax[i,j].imshow(img)
                    ax[i,j].spines['top'].set_visible(False)
                    ax[i,j].spines['left'].set_visible(False)
                    ax[i,j].spines['bottom'].set_visible(False)
                    ax[i,j].spines['right'].set_visible(False)
                    ax[i,j].set_xticks([])
                    ax[i,j].set_yticks([])
        plt.show()

    def pintar_estado(self):
        # Presenta gráficamente el mapa completo del juego con toda la información de la partida
        fig,ax = plt.subplots(self.n,self.m,figsize=(2,2))
        plt.subplots_adjust(wspace=0, hspace=0) # unes las figuras
        #ax[0,0].set_title('Cartas Jugador 1')
        for i in range(self.n):
            for j in range(self.m):
                with open("./Images/{}.JPG".format(int(self.estado[i,j])), 'rb') as f:
                    img = Image.open(f)
                    plt.axis('on')
                    ax[i,j].imshow(img)
                    ax[i,j].spines['top'].set_visible(False)
                    ax[i,j].spines['left'].set_visible(False)
                    ax[i,j].spines['bottom'].set_visible(False)
                    ax[i,j].spines['right'].set_visible(False)
                    ax[i,j].set_xticks([])
                    ax[i,j].set_yticks([])
        plt.show()

    def transicion(self,casilla,accion):
        # Casilla : ua tupla (x,y) en la cual se ejecuta la accion
        # Accion : 0 para destapar, 1 para poner bandera
        (x,y) = casilla
        if self.juego_activo:

            if accion == 0: #revelar casilla
                casilla_revelada = self.mapa[x,y]
                self.estado[x,y] = casilla_revelada
                #print(casilla_revelada)
                if casilla_revelada == 9:
                    print("Perdio")
                    self.juego_activo = False

            elif accion == 1: #poner bandera // marcas casillas
                if self.num_banderas > 0: #Verifica que tenga banderas que poner
                    self.estado[x,y] = 11
                    self.num_banderas = self.num_banderas -1
                else:
                    print("No puedes colocar mas banderas")
                    self.juego_activo = False

            elif accion ==2: #quitar bandera
                if self.estado[x,y]==11:
                    self.estado[x,y]=10
                    self.num_banderas = self.num_banderas +1
            else:
                print("Accion no permitida")
        else:
            print("Juego Inactivo")

        return self.estado

    def test_objetivo(self):
        for bomba in self.bombas:
            x,y = bomba
            if self.estado[x,y] != 11:
                return False
        return True
```

### Representación del mundo del Buscaminas en lógica proposicional

Debemos representar las letras proposicionales, las cuales cruzan la información sobre cada casilla. En nuestro problema esta información consistirá en: la columna (`x`), la fila (`y`), y el tipo de información. Definimos:

* `P(x, y, 10)` es verdadero sii en la casilla $(x,y)$ se sabe que esta tapada.
* `P(x, y, 0)` es verdadero sii el agente cree que en la casilla $(x,y)$ no existen casillas adyacentes con bombas.
* `P(x, y, 1)` es verdadero sii en la casilla $(x,y)$ se percibe 1 bomba adyacente a la casilla.
* `P(x, y, 2)` es verdadero sii en la casilla $(x,y)$ se percibe 2 bombas adyacentes a la casilla.
* `P(x, y, 3)` es verdadero sii en la casilla $(x,y)$ se percibe 3 bombas adyacentes a la casilla.
* `P(x, y, 4)` es verdadero sii en la casilla $(x,y)$ se percibe 4 bombas adyacentes a la casilla.
* `P(x, y, 5)` es verdadero sii en la casilla $(x,y)$ se percibe 5 bombas adyacentes a la casilla.
* `P(x, y, 6)` es verdadero sii en la casilla $(x,y)$ se percibe 6 bombas adyacentes a la casilla.
* `P(x, y, 7)` es verdadero sii en la casilla $(x,y)$ se percibe 7 bombas adyacentes a la casilla.
* `P(x, y, 8)` es verdadero sii en la casilla $(x,y)$ se percibe 8 bombas adyacentes a la casilla.
* `P(x, y, 9)` es verdadero sii en la casilla $(x,y)$ se percibe que hay una bomba.


```python
class codigos:
    
    # Clase para agrupar las funciones de codificación
    # de letras proposicionales
    
    def __init__(self, Nf, Nc, No):
        self.Nf = Nf # Número de filas
        self.Nc = Nc # Número de columnas
        self.No = No # Número de opciones de información
        
    def codifica(self, f, c, Nf, Nc):
        # Funcion que codifica la fila f y columna c
        assert((f >= 0) and (f <= Nf - 1)), 'Primer argumento incorrecto! Debe ser un numero entre 0 y ' + str(Nf) - 1  + "\nSe recibio " + str(f)
        assert((c >= 0) and (c <= Nc - 1)), 'Segundo argumento incorrecto! Debe ser un numero entre 0 y ' + str(Nc - 1)  + "\nSe recibio " + str(c)
        n = Nc * f + c
        # print(u'Número a codificar:', n)
        return n

    def decodifica(self, n, Nf, Nc):
        # Funcion que codifica un caracter en su respectiva fila f y columna c de la tabla
        assert((n >= 0) and (n <= Nf * Nc - 1)), 'Codigo incorrecto! Debe estar entre 0 y ' + str(Nf * Nc - 1) + "\nSe recibio " + str(n)
        f = int(n / Nc)
        c = n % Nc
        return f, c

    def P(self, f, c, o):
        # Funcion que codifica tres argumentos
        assert((f >= 0) and (f <= self.Nf - 1)), 'Primer argumento incorrecto! Debe ser un numero entre 0 y ' + str(Nf - 1) + "\nSe recibio " + str(f)
        assert((c >= 0) and (c <= self.Nc - 1)), 'Segundo argumento incorrecto! Debe ser un numero entre 0 y ' + str(Nc - 1) + "\nSe recibio " + str(c)
        assert((o >= 0) and (o <= self.No - 1)), 'Tercer argumento incorrecto! Debe ser un numero entre 0 y ' + str(No - 1)  + "\nSe recibio " + str(o)
        v1 = self.codifica(f, c, self.Nf, self.Nc)
        v2 = self.codifica(v1, o, self.Nf * self.Nc, self.No)
        codigo = chr(256 + v2)
        return codigo

    def Pinv(self, codigo):
        # Funcion que codifica un caracter en su respectiva fila f, columna c y objeto o
        x = ord(codigo) - 256
        v1, o = self.decodifica(x, self.Nf * self.Nc, self.No)
        f, c = self.decodifica(v1, self.Nf, self.Nc)
        return f, c, o
    
    def imprime_formula(self, fml):
        for c in fml:
            if c in ['Y', '>', '-']:
                print(c, end="")
            else:
                x, y, o = self.Pinv(c)
                if o == 10:
                    print("Esta tapada en " + str((x, y)), end="")
                if o == 0:
                    print("No hay bombas adyacentes " + str((x, y)), end="")
                if o == 1:
                    print("1 Bomba Adyacente alrededor de " + str((x, y)), end="")
                if o == 2:
                    print("2 Bombas Adyacentes alrededor de " + str((x, y)), end="")
                if o == 3:
                    print("3 Bombas Adyacentes alrededor de  "+ str((x, y)), end="")
                if o == 4:
                    print("4 Bombas Adyacentes alrededor de " + str((x, y)), end="")
                if o == 5:
                    print("5 Bombas Adyacentes alrededor de " + str((x, y)), end="")
                if o == 6:
                    print("6 Bombas Adyacentes alrededor de " + str((x, y)), end="")
                if o == 7:
                    print("7 Bombas Adyacentes alrededor de " + str((x, y)), end="")
                if o == 8:
                    print("8 Bombas Adyacentes alrededor de " + str((x, y)), end="")
                if o == 9:
                    print("Hay una bomba en" + str((x, y)), end="")
```

### Reglas Unicidad

Ahora usamos fórmulas de la lógica proposicional para representar aspectos del mundo. Comenzamos con una parte de la información sobre la relación entre el numero de bombas alrededor en una casilla y los caso que no deben suceder (unicidad) . Necesitamos incluir en la base de conocimiento las siguientes fórmulas:

1. $P(0,0,1) \to \bigl(\neg P(0,0,2)\wedge... P(0,0,8)\bigr)$ *(si hay una bomba adyacente en $(0,0)$, no hay 1,2,3,4,5,6,7,8 bombas $(0,0)$)*
* $P(1,1,2) \to \bigl( P(0,1,1)\wedge P(2,1,1)\wedge P(1,0,1)\wedge P(1,2,1)\bigr)$ *(si hay dos bomba adyacentes en $(1,1)$, no hay 1,3,4,5,6,7,8,9 bombas)*
* $\vdots$

Con el fin de implementar los algoritmos de búsqueda, usaremos fórmulas en forma 'pseudo' cláusulas de Horn (ver diapositivas sesión 7), por lo que reescribimos las anteriores fórmulas así:

*Conjunto de fórmulas que expresan la fórmula (1) arriba:*
* $\neg P(0,0,1)\to \neg P(0,0,2)$
* $\neg P(0,0,1)\to \neg P(0,0,3)$


<img src="./Images/regla_u.png" width="700px">

```python
def forms_uni(B):
    #Generamos reglas de unicidad
    formulas_unicidad =[]
    for x in range(B.n):
        for y in range(B.m):
            for i in range(10):
                for j in range(10):
                    if i != j:
                        formula = "{}>-{}".format(cods.P(x,y,i),cods.P(x,y,j))
                        formulas_unicidad.append(formula)
    return formulas_unicidad

print(len(forms_uni(B)))
```

### Reglas reconocimiento no bombas

Ahora definimos las formulas para que el jugador elimine casos donde clasifique correctamente donde se encuentra una bomba.

1. $P(0,0,1)\wedge \neg P(0,1,9) \wedge \neg P(1,0,9) \to P(1,1,9)$ *(si hay una bomba adyacente en $(0,0)$ y no hay bombas adyacentes en $(0,1)$ y $(1,0)$, entonces ahi una bomba adyacente en $(1,1)$)*

La ecuacion de arriba es un caso simple, sin embargo suceden casos mas complicados cuando hay dos bombas adyacentes, pero la idea sigue siendo la misma por ejemplo 

1. $P(0,0,2)\wedge \neg P(0,1,9) \to P(1,1,9)$ *(si hay dos bomba adyacente en $(0,0)$ y no hay bombas en $(0,1)$ , entonces hay una bomba en $(1,1)$)*
1. $P(0,0,2)\wedge \neg P(0,1,9) \to P(1,0,9)$ *(si hay dos bombas adyacentes en $(0,0)$ y no hay bomba en $(0,1)$ , entonces hay una bomba en $(1,0)$)*


<img src="./Images/regla_b.png" width="200px">

```python
def forms_num_nobomb(B):
    formula_numeros = []
    for x in range(B.n):
        for y in range(B.m):
            ady = adyacentes((x,y),B.n,B.m) #Saca las casillas adyacentes de la casilla (x,y)
            for a in ady:
                formula_numeros.append("{}>-{}".format(cods.P(x,y,0),cods.P(a[0],a[1],9)))
            for i in range(1,9):
                if len(ady) >= i:
                    combinaciones = combinations(ady,i)
                    for combinacion in combinaciones:
                            tmp_regla=['-'+cods.P(indice[0],indice[1],9) for indice in ady if indice not in combinacion]
                            regla = cods.P(x,y,i)
                            for coso in tmp_regla:
                                regla += 'Y'+coso
                            tmp_regla = [cods.P(indice[0],indice[1],9) for indice in combinacion]
                            for coso in tmp_regla:
                                formula_numeros.append(regla+'>'+coso)
    return formula_numeros

print(len(forms_num_nobomb(B)))
```

### Regla reconocimiento bombas

Ahora definimos las formulas para que el jugador sepa donde no hay una bomba o se encuentre una bomba utililizando la base de conocimiento.

1. $P(0,0,1)\wedge \neg P(0,1,9) \wedge \neg P(1,0,9) \to P(1,1,9)$ *(si hay una bomba adyacente en $(0,0)$ y hay una bomba adyacentes en $(0,1)$ entonces no hay bombas en $(1,1)$ y $(1,0)$)*

La ecuacion de arriba es un caso simple, sin embargo suceden casos mas complicados cuando hay dos bombas adyacentes, pero la idea sigue siendo la misma.

<img src="./Images/regla_nb.png" width="200px">

```python
def forms_num_bomb(B):
    formula_numeros = []
    for x in range(B.n):
        for y in range(B.m):
            ady = adyacentes((x,y),B.n,B.m) #Saca las casillas adyacentes de la casilla (x,y)
            for i in range(1,9):
                if len(ady) >= i:
                    combinaciones = combinations(ady,i)
                    for combinacion in combinaciones:
                            tmp_regla = [cods.P(indice[0],indice[1],9) for indice in combinacion]
                            regla = cods.P(x,y,i)
                            for coso in tmp_regla:
                                regla += 'Y'+coso
                            tmp_regla=['-'+cods.P(indice[0],indice[1],9) for indice in ady if indice not in combinacion]
                            for coso in tmp_regla:
                                formula_numeros.append(regla+'>'+coso)
    return formula_numeros

print(len(forms_num_bomb(B)))
```

```python
class lp_query:

    def __init__(self, base_conocimiento_lista, cods):

        # Con base en la lista de cláusulas en la base de
        # conocimiento, crea un diccionario optimizado para
        # la búsqueda
        # Input: base_conocimiento_lista, que es una lista de fórmulas 'pseudo' cláusulas de Horn
        #        cods, un objeto de clase codigos
        self.base_conocimiento = {'datos':[], 'reglas':{}}
        for formula in base_conocimiento_lista:
            indice_conectivo = formula.find('>')
            if indice_conectivo > 0:
                cuerpo = formula[:indice_conectivo]
                cabeza = formula[indice_conectivo + 1:]
                try:
                    self.base_conocimiento['reglas'][cabeza].append(cuerpo)
                except:
                    self.base_conocimiento['reglas'][cabeza] = [cuerpo]
            else:
                try:
                    self.base_conocimiento['datos'].append(formula)
                except:
                    self.base_conocimiento['datos'] = [formula]
        self.cods = cods

    def visualizar(self, parte='todo'):
        if (parte=='todo') or (parte=='datos'):
            print("Datos:")
            for l in self.base_conocimiento['datos']:
                self.cods.imprime_formula(l)
                print('')
        if (parte=='todo') or (parte=='reglas'):
            print("\nReglas:")
            for k in self.base_conocimiento['reglas'].keys():
                for c in self.base_conocimiento['reglas'][k]:
                    self.cods.imprime_formula(c)
                    print(">", end="")
                    self.cods.imprime_formula(k)
                    print(" ")

    def reglas_aplicables(self, head):
        # Devuelve una lista de cláusulas de Horn cuya cabeza
        # es el estado/literal
        # Input: head, que es un literal
        # Output: lista de cláusulas de Horn
        try:
            cuerpo = self.base_conocimiento['reglas'][head]
        except:
            cuerpo = []

        return cuerpo

    def transicion(self, head, body):
        # Devuelve una lista con los literales en la clausula
        # Input: head, que es un literal se asume que head:- body es una regla
        #        body, que es una lista de cláusulas de Horn
        return body.split('Y')

    def test_objetivo(self, literal):
        # Devuelve True/False dependiendo si el literal
        # está en la base de datos
        # Input: literal
        # Output: True/False
        literales = self.base_conocimiento['datos']
        return True if literal in literales else False
```

Llenamos de conocimiento la base

```python
def TELL(base, formula):
    indice_conectivo = formula.find('>')
    if indice_conectivo > 0:
        cuerpo = formula[:indice_conectivo]
        cabeza = formula[indice_conectivo + 1:]
        try:
            if cuerpo not in base.base_conocimiento['reglas'][cabeza]:
                base.base_conocimiento['reglas'][cabeza].append(cuerpo)
        except:
            base.base_conocimiento['reglas'][cabeza] = [cuerpo]
    else:
        lista_hechos = formula.split('Y')
        for literal in lista_hechos:
            try:
                if literal not in base.base_conocimiento['datos']:
                    base.base_conocimiento['datos'].append(literal)
            except:
                base.base_conocimiento['datos'] = [literal]
```

Luego para que funcionara correctamente nuestra reglas decidimos modificar la funcion ask para que pudiera resolver nuestro problema eficientemente.

```python
def ASK(objetivo,valor,base):
    #objetivo es un literal positivo
    return and_or_grap_search(objetivo,valor,base)
    
def and_or_grap_search(objetivo,valor,base):
    return or_search(objetivo,base,valor,[])

def or_search(head,base,valor,camino):
    if '-' not in head and base.test_objetivo(head):
        return valor == True
    elif '-' in head and base.test_objetivo(head):
        return valor == False
    elif head in camino :
        return False
    posibles_reglas = base.reglas_aplicables(head)
    for regla in posibles_reglas:
        camino1 = [head] + camino
        resultado = and_search(base.transicion(head,regla),base,camino1,valor)
        if resultado == True :
            return True #Suceso
    return False #perdida

def and_search(literales,base,camino,valor):
    for head in literales:
        if '-' in head:
            plan = or_search(head,base,False,camino)
        else:
            plan = or_search(head,base,True,camino)
        if not plan:
            return False
    return True
```

```python
def jugar(B,Base_buscaminas):
    aleatorio1 = rd.randint(0,(B.n)-1)
    aleatorio2 = rd.randint(0,(B.m)-1)
    # Juega aleatoriamente el primero
    print("Empezaré destapando ({},{})".format(aleatorio1,aleatorio2))
    B.transicion((aleatorio1,aleatorio2),0)
    B.pintar_estado()
    TELL(Base_buscaminas,cods.P(aleatorio1,aleatorio2,B.estado[aleatorio1,aleatorio2]))
    while B.juego_activo:
        modificacion = False
        for x in range(B.n):
            for y in range(B.m):
                if (B.estado[x,y]==10):
                    print("Estoy determinando la casilla ({},{})".format(x,y))
                    if ASK(cods.P(x,y,9),True,Base_buscaminas):
                        print("Pude determinar que en la casilla ({},{}) hay bomba, pongo bandera".format(x,y))
                        B.transicion((x,y),1)
                        modificacion = True
                        B.pintar_estado()
                    if ASK('-'+cods.P(x,y,9),True,Base_buscaminas):
                        print("Pude determinar que en la casilla ({},{}) no hay bomba, destapo".format(x,y))
                        B.transicion((x,y),0)
                        TELL(Base_buscaminas,cods.P(x,y,B.estado[x,y]))
                        B.pintar_estado()
                        modificacion = True
        #Ya recorrió todas las casillas tapadas
        if B.test_objetivo():
            print('Ganó!')
            return B
        if not modificacion:
            print("No pudo determinar nada, entonces eligió aleatoriamente")
            aleatorio1 = rd.randint(0,(B.n)-1)
            aleatorio2 = rd.randint(0,(B.m)-1)
            while (B.estado[aleatorio1,aleatorio2] != 10):
                aleatorio1 = rd.randint(0,(B.n)-1)
                aleatorio2 = rd.randint(0,(B.m)-1)
            print("Destaparé ({},{})".format(aleatorio1,aleatorio2))
            B.transicion((aleatorio1,aleatorio2),0)
            TELL(Base_buscaminas,cods.P(aleatorio1,aleatorio2,B.estado[aleatorio1,aleatorio2]))
            B.pintar_estado()
    B.pintar_tablero()
    return B
```
