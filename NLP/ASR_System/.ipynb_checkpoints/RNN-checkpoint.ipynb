# Funciones y Diccionarios Útiles

```python
# Traduce letras a fonemas
letter2phonem = {
    'a':'a','b':'b','v':'b','w':'b','c':'k','z':'0',
    'ch':'ch','d':'d','e':'e','f':'f','g':'g','gu':'g',
    'h':'','i':'i','y':'i','g':'j','j':'j','ca':'k',
    'cc':'k','ci':'s','ce':'s','co':'k','cu':'k','qu':'k',
    'k':'k','l':'l','ll':'ll','m':'m','n':'n','ñ':'ñ',
    'o':'o','p':'p','r':'r','s':'s','t':'t','u':'u',
    'x':'gs','y':'y',' ':' '
}

# Traduce fonemas a indices
phonem2index = {
    'e':0,'s':1,'t':2,'u':3,'d':4,'i':5,'k':6,'a':7,'ch':8,
    'l':9,'o':10,'m':11,'n':12,'f':13,'r':14,'p':15,'':16,' ':17
}

# Traduce indices a fonemas
index2phonem = {
    0:'e',1:'s',2:'t',3:'u',4:'d',5:'i',6:'k',7:'a',8:'ch',
    9:'l',10:'o',11:'m',12:'n',13:'f',14:'r',15:'p',16:'',17:' '
}
```

```python
def getLetters(word,dic):
    '''
        Funcion que dada una palabra (word) y un 
        diccioario (dic), retorna una lista con 
        las letras que componen la palabra y las
        combinaciones de letras que forman fonemas.
        
        INPUT:
            word: string
            dic: diccioario
        OUTPUT: lista de strings
    '''
    special = [n for n in dic if len(n)>1]
    if(len(word)==1 or len(word)==0):
        return list(word) if word!='h' else []
    else:
        i = next((word.index(n) for n in special if n in word),None)
        if(i is None):
            return list(word)
        else:
            return getLetters(word[0:i],dic)+[word[i:i+2]]+getLetters(word[i+1:],dic)

def getPhonems(word,dic):
    '''
        Funcion que dada una palabra (word) y 
        un diccionario (dic) retorna una lista 
        con los fonemas que componen la palabra.
        
        INPUT:
            word: string
            dic: diccionario
        OUTPUT: lista de strings
    '''
    return [dic[n] for n in getLetters(word,dic)]
```

```python
def getFonemas(mel,n_fonema):
    '''
        Funcion que dado un espectrograma de mel y
        una cantidad de fonemas retorna una lista, 
        de longitud n_fonemas, de vectores los cuales 
        representan los fonemas que componen a la palbra
        
        INPUT:
            mel: matriz de 39x63
            n_foenma: enetro
        OUTPUT: lista de vectores longitud n_fonema
    '''
    rou = lambda x,n : int(x/n) if x%n == 0 else int(x/n)+1
    d = np.sqrt(sum(np.power(mel[1:]-mel[0:-1],2)))
    m = np.mean(d)
    index = [i for i,n in enumerate(d) if n>m]
    delta = rou(index[-1]-index[0],n_fonema)
    fonemas = []
    for f in range(n_fonema):
        I_1 = index[0]+(delta*f)
        I_2 = index[0]+(delta*(f+1))+1
        fonemas.append(mel[:,I_1:I_2])
    return fonemas
```

# Calculando Mel-Spectrograma

Calculamos los coeficientes mel

```python
# Creando la matriz del espectrograma
fs = 8000
n_window = int(0.02*fs)
len_over = int(0.01*fs) # Number overlap between samples
nfft = 512              # Number Points FFT
n_mels = 39

lista_espectrogramas_mel =[]
for sonido in lista_sonidos:
    mel = librosa.feature.melspectrogram(sonido[0],sr=fs,n_fft=512,win_length=n_window,window='hann',n_mels=n_mels)
    lista_espectrogramas_mel.append((mel,sonido[1]))
```

# Dividiendo Data-Set en train y test

```python
# Lista de los espectrogramas
X = [lista_espectrogramas_mel[i][0] for i in range(len(lista_espectrogramas_mel))]
```

```python
# Matriz representacion de secuencias de palabras
I = np.concatenate([np.eye(18)],axis=0)
Y = [I[:,[phonem2index[j] for j in getPhonems(word[2],letter2phonem)]] for word in lista_sonidos]
```

```python
# Ejemplo
Y[0]
```

Para que todas las matrices de Y tengan las mismas dimensiones de (18,16) se le agregan a todas las matrices los vectores columnas faltantes. Esto se hace ya que se quiere que las matrices tengan las mismas dimensiones para la red y se define con 16 ya que la palabra del modelo que más fonemas tiene posee 16 fonemas.

```python
# Agregando vectores columnas
for i,w in enumerate(Y):
    c = 16-w.shape[1]
    Y[i] = np.concatenate([Y[i],np.zeros((18,c))],axis=1)
```

### Dividiendo en Train Test

```python
# Dividiendo en train y test
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Pasando a array de numpy (formato necesario para tensorflow)
X_train= np.array([m.T for m in X_train])
X_test= np.array([m.T for m in X_test])
Y_train= np.array([m.T for m in Y_train])
Y_test= np.array([m.T for m in Y_test])
```

# Modelo de Tensorflow

La arquitectura de la red está basada de la propuesta en https://github.com/keras-team/keras/issues/6063, además de le agregaron dos capas densas (Dense) ya que se observó que mejoraba los resultados. Poro otro lado no se normalizaron los datos ya que tras varias pruebas los mejores resultados se lograban con los datos sin normalizar y además fue necesario que la cantidad de épocas (epochs) fueran 1000 y el uso de 210 grabaciones para obtener un accuracy de 0,66, a diferencia del 0,12 que se lograba solo con 10 grabaciones.

```python
# Dimensiones de los Train-test
print("X_train shape: ",X_train.shape)
print("Y_train shape: ",Y_train.shape)
print("X_test shape: ",X_test.shape)
print("Y_test shape: ",Y_test.shape)
```

```python
# Arquitectura red neuronal
model = Sequential()  
model.add(LSTM(18,input_dim=39, return_sequences=False))  

model.add(Dense(18, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(18, activation='relu'))
model.add(Dropout(0.1)) 

model.add(tf.keras.layers.RepeatVector(16))
model.add(LSTM(18, return_sequences=True))  
model.add(tf.keras.layers.TimeDistributed(Dense(18)))
model.add(tf.keras.layers.Activation('linear'))   
model.compile(loss='mean_squared_error', optimizer='rmsprop', metrics=['accuracy'])  
```

```python
model.fit(X_train,Y_train,epochs=2000,validation_data=(X_test,Y_test))
```

### Realizando la predicción

Ahora vamos a realizar las predicciones sobre el conjunto de X_train utilizando el modelo anteriormente entrenado

¿Qué modelo funciona mejor? ¿Cúal es la razón para esto?
- En nuestro caso el HMM funciona mejor. Esto lo atribuimos al hecho de que la red necesita muchos más datos para obtener mejores resultados que el planteado con las cadenas de markov. De esta forma esperariamos que al ingresar más datos de entrenamiento la red mejore.

¿Discuta sobre las ventajas y desventajas del modelo basado en HMM?

Las desventajas serían que dado que este modelo se contruye de manera manual, si quisieramos tener un procesamiento del lenguaje más grande, necesitaríamos más tiempo entrenando el modelo lo cual podría tomar muchos años (en el peor caso).

Las desventajas serían que es muy dificil saber qué está hciendo el modelo por detras y no se podría saber si el modelo está realizando sesgos sobre la voz, discriminación de voz, etc.
